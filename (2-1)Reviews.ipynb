```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
brands=[]
products = []
colors = []
reviews=[]
dates=[]
driver = webdriver.Chrome('C:\\Users\\CE206-38\\Downloads\\chromedriver_win32\\chromedriver')

# for n in range(0,len(url)):
#     driver.get('https://www.etudehouse.com/kr/ko/product/detail?onlineProdSn='+str(url[n]))
#     time.sleep(1)
driver.get('https://www.etudehouse.com/kr/ko/product/detail?onlineProdSn=6578')

xpath = """//*[@id="ap_container"]/div/div/div[3]/div[1]/div[1]/ul/li[2]/button"""
driver.find_element_by_xpath(xpath).click()
time.sleep(1)

brands=[]
products = []
colors = []
reviews=[]
dates=[]

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
brand="에뛰드하우스"
product=soup.find('h3',class_="h_title").get_text()
flag = True
while flag:
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    size = len(soup.find('div','pagination ＠paging-apply').find_all('a'))
    size = size-3
    print("size:", size)
    if size==10:
        for _ in range(size+1):
            html = driver.page_source
            soup = BeautifulSoup(html,'html.parser')
            for re in range(len(soup.find('div', 'review_list').find_all('dl','review'))):
                brands.append(brand)
                products.append(product)
                color_tmp=soup.find_all('small',class_='opt')
                color_tmp=color_tmp[re].get_text()
                color_tmp=color_tmp.replace('옵션:','')
                colors.append(color_tmp)
                date=soup.find_all('dl',class_="review")[re].find('small').get_text()
                dates.append(date)
                review=soup.find_all('div',class_="text reduce")
                review_tmp=review[re].get_text()
                review_tmp=review_tmp.replace('\n','')
                reviews.append(review_tmp)

            time.sleep(0.5)
            xpath='//*[@id="ap_container"]/div/div/div[3]/div[1]/div[2]/div[2]/div[3]/div/a[12]'
            driver.find_element_by_xpath(xpath).click()

    else:
        for _ in range(size-1):
            html = driver.page_source
            soup = BeautifulSoup(html,'html.parser')
            for re in range(len(soup.find('div', 'review_list').find_all('dl','review'))):
                brands.append(brand)
                products.append(product)
                color_tmp=soup.find_all('small',class_='opt')
                color_tmp=color_tmp[re].get_text()
                color_tmp=color_tmp.replace('옵션:','')
                colors.append(color_tmp)
                date=soup.find_all('dl',class_="review")[re].find('small').get_text()
                dates.append(date)
                review=soup.find_all('div',class_="text reduce")
                review_tmp=review[re].get_text()
                review_tmp=review_tmp.replace('\n','')
                reviews.append(review_tmp)
            time.sleep(0.5)
            xpath='//*[@id="ap_container"]/div/div/div[3]/div[1]/div[2]/div[2]/div[3]/div/a['+str(size+2)+']'
            driver.find_element_by_xpath(xpath).click() 
        flag=False
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
for re in range(len(soup.find('div', 'review_list').find_all('dl','review'))):
    brands.append(brand)
    products.append(product)
    color_tmp=soup.find_all('small',class_='opt')
    color_tmp=color_tmp[re].get_text()
    color_tmp=color_tmp.replace('옵션:','')
    colors.append(color_tmp)
    date=soup.find_all('dl',class_="review")[re].find('small').get_text()
    dates.append(date)
    review=soup.find_all('div',class_="text reduce")
    review_tmp=review[re].get_text()
    review_tmp=review_tmp.replace('\n','')
    reviews.append(review_tmp)
    time.sleep(0.5)
len(reviews)
```
