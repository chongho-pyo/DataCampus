## MAC 립스틱
- 립스틱/비바 글램 25는 rgb 추출 불가

```python
html = requests.get('https://www.maccosmetics.co.kr/products/13854/makeup')
soup = BeautifulSoup(html.text, 'html.parser')
product_urls = soup.find_all('a', 'product__name-link js-product__link-to-spp product_header_name')
```

```python
product_urls
```

```python
names = []
prices = []
colors = []
rgbs = []
for data in product_urls:
    url = "https://www.maccosmetics.co.kr" +  data['href']
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    #tag, class 순서로 그냥 써도 됨
    tmp= soup.find('ul', 'shade-picker__colors')
    
    if not tmp:
        continue
    
    for color, rgb in zip(tmp.find_all('div', 'shade-picker__color-name'),tmp.find_all('div', 'shade-picker__color-chip')):
        names.append(soup.find('h3', 'product__subline').get_text())
        prices.append(soup.find('span', 'product__price--standard').get_text()[2:])
        colors.append(color.get_text())
        rgbs.append(rgb['style'].split(' ')[1][:-1])
        
names[0], prices[0], colors[0], rgbs[0] 
```

```python
mac = pd.DataFrame({'상품명' : names,
                 '가격' : prices,
                 '색상': colors,
                  'rgb': rgbs}, columns=['상품명', '가격', '색상', 'rgb'])
mac
```

## 에스티로더

```python
html = requests.get('https://www.esteelauder.co.kr/products/649')
soup = BeautifulSoup(html.text, 'html.parser')
product_urls = soup.find_all('a','product_brief__image-container')
```

```python
names = []
prices = []
colors = []
rgbs = []
for data in product_urls:
    url = "https://www.esteelauder.co.kr" +  data['href']
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    tmp = soup.find('ul','js-shade-picker shade-list')
    for rgb in tmp.find_all('li', 'swatches--single'):
        if soup.find('h3','product-full__title product_brief__sub-line'):
            names.append(soup.find('h3','product-full__title product_brief__sub-line').get_text())
        else:
            names.append(soup.find('h4','product_brief__sub-header').get_text())
        prices.append(soup.find('div','product-full__price-text').get_text().split('\n')[1].lstrip()[1:])
        rgbs.append(rgb.find('div', 'swatch--1')['style'].split(':')[1])
        colors.append(rgb.find('a', 'swatch swatch--selected')['name'])
names[0], prices[0], colors[0], rgbs[0]        
```

```python
estee = pd.DataFrame({'상품명' : names,
                 '가격' : prices,
                 '색상': colors,
                  'rgb': rgbs}, columns=['상품명', '가격', '색상', 'rgb'])
estee
```

## Nars

```python
from selenium import webdriver
import time
driver = webdriver.Chrome('../Driver/chromedriver_win32/chromedriver')
driver.get("https://www.narscosmetics.co.kr/ko/lipstick")

for _ in range(0,18):
    driver.execute_script("window.scrollBy(0, 700);")
    time.sleep(1)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
names = []
prices = []
colors = []
rgbs = []

for data in soup.find_all('div', 'product-tile-info clearfix'):
    if not data.find('div','selected-shade-name'):
        continue
    names.append(data.find('a', 'name-link').get_text().split('\n')[1])
    prices.append(data.find('span', 'price-sales').get_text().split('\n')[1][1:])
    colors.append(data.find('div','selected-shade-name').get_text().split('\n')[1])
    rgbs.append(data.find('div', 'swatch-block-plp')['style'].split(' ')[1])
    
names[0], prices[0], colors[0], rgbs[0]
```

```python
nars = pd.DataFrame({'상품명' : names,
                 '가격' : prices,
                 '색상': colors,
                  'rgb': rgbs}, columns=['상품명', '가격', '색상', 'rgb'])
nars
```

###  입생로랑

```python
html = requests.get('https://www.yslbeautykr.com/ko_KR/makeup/lips/lipstick')
soup = BeautifulSoup(html.text, 'html.parser')
product_urls = soup.find_all('a','product_image_wrapper b-product_img-link')
```

```python
def rgb_get(url):
    img = Image.open(urlopen(url))
    img = img.convert("RGBA")
    rgb = img.getpixel((10, 10))
    
    r = str(hex(rgb[0]).replace('0x','')).zfill(2)
    g = str(hex(rgb[1]).replace('0x','')).zfill(2)
    b = str(hex(rgb[2]).replace('0x','')).zfill(2)
    return r+g+b
```

```python
names = []
prices = []
colors = []
rgbs = []
for url in product_urls:
    html = requests.get(url['href'])
    soup = BeautifulSoup(html.text, 'html.parser')
    col = soup.find('ul','swatches js_swatches color contentcarousel_list')
    for img in col.find_all('img','swatch_image_color'):
        names.append(soup.find('h1',class_='product_name').get_text().replace('\n', '').replace('\t', ''))
        prices.append(soup.find('p','product_price price_sale b-product_price-sale b-product_price').get_text().replace('\n','')[:-1])
        colors.append(soup.find('span','swatch_text_color').get_text())
        rgbs.append(rgb_get(img['src']))
                
names[0], prices[0], colors[0], rgbs[0] 
```

```python
ysl = pd.DataFrame({'상품명' : names,
                 '가격' : prices,
                 '색상': colors,
                  'rgb': rgbs}, columns=['상품명', '가격', '색상', 'rgb'])
ysl
```
